```python
def recursive_sum(num):
    if num==0:
        return 0
    result = num + recursive_sum(num-1)
    return result
```

```python
recursive_sum(6)
```

```python
def factorial(num):
    if num<0:
        print("Sorry, it's negative")
        return
    if num==0:
        return 1
    return num*factorial(num-1)
```

```python
factorial(6)
```

```python
def fibonacci(num, fib_ser):
    if num<2:
        return
    l=len(fib_ser)
    new_num = fib_ser(l-1) + fib_ser(l-2)
    fib_ser.append(new_num)
    fibonacci(num-1,fib_ser)
```

```python
series=[0,1]
fibonacci(10,series)
```
